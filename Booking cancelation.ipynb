```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
def tuning(cls,parameter,X_train,y_train):
    cv = StratifiedKFold(n_splits=5,shuffle=True)
    cls_grid = GridSearchCV(cls,parameter,cv=cv,scoring='f1',n_jobs=-1)
    cls_grid.fit(X_train,y_train)
    best_hp = cls_grid.best_params_
    return cls_grid.best_estimator_,best_hp
                         
```

```python
dt = DecisionTreeClassifier()
```

```python
dt_parameter_grid = {
    'criterion':['gini','entropy'],
    'max_depth':[12,13,14]
}
```

```python
tuning(dt,dt_parameter_grid,X_train,y_train)
```

```python
dt = DecisionTreeClassifier(criterion= 'gini', max_depth= 14)
```

```python
dt.fit(X_train,y_train)
```
