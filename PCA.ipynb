```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('AmesHousing.csv')
```

```python
data
```

```python
data.info()
```

```python
data.drop(columns=['Order','PID'],inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
num = data.select_dtypes(include=['int64','float64']).columns
```

```python
num
```

```python
cat = data.select_dtypes(include=['object']).columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='median')
```

```python
data[num] = si.fit_transform(data[num])
```

```python
data
```

```python
si = SimpleImputer(strategy='most_frequent')
data[cat] = si.fit_transform(data[cat])
```

```python
si
```

```python
data.isna().sum().values
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
ssc = StandardScaler()
for i in cat:
    data[i] = le.fit_transform(data[i])
```

```python
data[num] = ssc.fit_transform(data[num])
```

```python
data['Alley'].unique()
```

```python
X = data.drop('Alley',axis=1)
y = data['Alley']
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
```

```python
X = pca.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
with open("PCA_model.pkl", "wb") as file:
    pickle.dump(PCA, file)

print("✅ Model saved!")
```
