```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Base URL of the site
base_url = "https://quotes.toscrape.com/page/{}/"
```

```python
# List to store all scraped data
all_quotes = []
```

```python
# Loop through multiple pages (you can increase the range if needed)
for page in range(1, 6):  # Scrapes pages 1 to 5
    url = base_url.format(page)
    response = requests.get(url)

    # Stop loop if page is not found
    if response.status_code != 200:
        break

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    quotes = soup.find_all('div', class_='quote')

    # Extract quote text and author
    for quote in quotes:
        text = quote.find('span', class_='text').get_text()
        author = quote.find('small', class_='author').get_text()
        all_quotes.append({'Quote': text, 'Author': author})
```

```python
# Convert to DataFrame
df = pd.DataFrame(all_quotes)
```

```python
# Save to CSV
df.to_csv('quotes.csv', index=False)

print("Scraping completed. Data saved to 'quotes.csv'")
```
